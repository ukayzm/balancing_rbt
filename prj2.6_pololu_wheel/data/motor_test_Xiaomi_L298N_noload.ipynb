```python
import numpy as np
import pandas as pd
from scipy import interpolate
import matplotlib.pyplot as plt
```

```python
def plot_graph(title, f1, f2):
    # read input file
    data1 = pd.read_csv(f1)
    data2 = pd.read_csv(f2)

    # interpolate
    f1 = interpolate.interp1d(data1.pwm, data1.RPM)
    #f1 = interpolate.interp1d(data1.PWM, data1.average_RPM, kind='cubic')
    f2 = interpolate.interp1d(data2.pwm, data2.RPM)
    x1 = np.linspace(0, 260, num=261, endpoint=True)
    x2 = np.linspace(0, 260, num=261, endpoint=True)

    # plot
    plt.plot(data1.pwm, data1.RPM, 'o', x1, f1(x1), '-', data2.pwm, data2.RPM, 'x', x2, f2(x2), '--')
    plt.xlabel('PWM')
    plt.ylabel('RPM')
    plt.grid(True)
    plt.legend(['', 'increasing', '', 'decreasing'])
    plt.title(title)
    plt.show()
```
