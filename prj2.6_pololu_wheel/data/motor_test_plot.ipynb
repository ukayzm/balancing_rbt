```python
import numpy as np
import pandas as pd
from scipy import interpolate
import matplotlib.pyplot as plt
```

```python
# read input file
data_filename1 = "Xiaomi_L298N_30Hz.dat.1"
data_filename2 = "Xiaomi_L298N_30Hz.dat.2"
data1 = pd.read_csv(data_filename1)
data2 = pd.read_csv(data_filename2)
data1.columns
```

```python
f1 = interpolate.interp1d(data1.PWM, data1.average_RPM)
#f1 = interpolate.interp1d(data1.PWM, data1.average_RPM, kind='cubic')
f2 = interpolate.interp1d(data2.PWM, data2.average_RPM)
x1 = np.linspace(0, 260, num=261, endpoint=True)
x2 = np.linspace(0, 260, num=261, endpoint=True)
plt.plot(data1.PWM, data1.average_RPM, 'o', x1, f1(x1), '-', data2.PWM, data2.average_RPM, 'x', x2, f2(x2), '--')
plt.xlabel('PWM')
plt.ylabel('RPM')
plt.grid(True)
plt.legend(['', 'increasing', '', 'decreasing'])
plt.show()
```
